```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd



file_path = 'modified_output.xlsx'  
df = pd.read_excel(file_path)

Query_data = df['Query'].head(10).tolist()
print(Query_data)
Query_data = [Query_data[:-4] if len(Query_data) > 2 else '' for Query_data in Query_data]
print(Query_data)

labels = Query_data


sf1_data = df['1GB'].head(10).tolist()

sf1_data = [round(item, 5) for item in sf1_data]
print(sf1_data)

sf3_data = df['3GB'].head(10).tolist()
sf3_data = [round(item, 5) for item in sf3_data]
print(sf3_data)

sf5_data = df['5GB'].head(10).tolist()
sf5_data = [round(item, 5) for item in sf5_data]
print(sf5_data)

sf7_data = df['7GB'].head(10).tolist()
sf7_data = [round(item, 5) for item in sf7_data]
print(sf7_data)




x = np.arange(len(labels))  
width = 0.2  

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - 1.5*width, sf1_data, width, label='sf=1', color='r')
rects2 = ax.bar(x - 0.5*width, sf3_data, width, label='sf=3', color='g')
rects3 = ax.bar(x + 0.5*width, sf5_data, width, label='sf=5', color='b')
rects4 = ax.bar(x + 1.5*width, sf7_data, width, label='sf=7', color='y')


def autolabel(rects, ax,fontsize=10):
 
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')



ax.set_ylabel('Execution Time(Seconds)')
ax.set_xlabel('Query')
ax.set_title('Exectution Time from Query1 to Query10 for the 4 scale factors')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


fig.tight_layout()

plt.show()
```

```python
import pandas as pd
import numpy as np
# Excel
file_path = 'modified_output.xlsx' 
df = pd.read_excel(file_path)


query = df['Query'].iloc[0:103].tolist()

query = [query[:-4] if len(query) > 2 else '' for query in query]
query = [query[6:] if len(query) > 2 else '' for query in query]
print(query)



values = df['1GB'].iloc[0:103].tolist()

print(query)
print(values)

y = query
x = values


fig, ax = plt.subplots(figsize=(20, 30)) 

ax.barh(y, x)
bars = ax.barh(y, x,color='blue')

# '--'
avg_value = np.mean(values)
ax.axvline(avg_value, color='red', linestyle='--', label=f'Average: {avg_value:.5f}')
ax.legend()  # 

# 
ax.text(avg_value + 0.01 * max(values), 105, f'Average: {avg_value:.5f}', color='red', va='center')

# 
for bar in bars:
    width = bar.get_width()
    ax.text(width + 0.01 * max(values),  # x position
            bar.get_y() + bar.get_height()/2,  # y position
            f'{width:.5f}',  # text
            va='center', ha='left', color='black', fontsize=9)

ax.xaxis.tick_top()
ax.yaxis.tick_left()

ax.set_xlabel('Scale1', position=(0.5, 1.02))
ax.set_ylabel('Query Number', position=(0, 1.02))
ax.set_title('Execution Time(Seconds)', position=(0.5, 1.05))

ax.set_yticks(np.arange(0, 102))

plt.tight_layout()  
plt.show()
```
