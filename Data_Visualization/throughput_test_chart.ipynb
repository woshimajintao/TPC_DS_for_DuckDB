```python
import numpy as np
import matplotlib.pyplot as plt


# 


def autolabel(rects, ax,fontsize=10):
    """"""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

sc_1_tp_1_time = 10.050887823104858
sc_1_tp_2_time = 10.481513023376465
sc_3_tp_1_time = 25.980229139328003
sc_3_tp_2_time = 26.450099229812622
sc_5_tp_1_time = 43.38998198509216
sc_5_tp_2_time = 44.39446187019348
sc_7_tp_1_time = 65.82485914230347
sc_7_tp_2_time = 61.08054518699646   

labels = ['1', '3', '5', '7']
data1 = [sc_1_tp_1_time, sc_3_tp_1_time, sc_5_tp_1_time,sc_7_tp_1_time] 
data1 = [round(num, 2) for num in data1]
data2 = [sc_1_tp_2_time, sc_3_tp_2_time, sc_5_tp_2_time,sc_7_tp_2_time]
data2 = [round(num, 2) for num in data2]


x = np.arange(len(labels))  
width = 0.35  

# 
#plt.figure(figsize=(10, 10))  # 105
fig, ax = plt.subplots(figsize=(10, 6)) 
# y
ax.set_ylim(0, 80)  # 060


data1 = ax.bar(x - width/2, data1, width, label='Throughput 1')
data2 = ax.bar(x + width/2, data2, width, label='Throughput 2')


ax.set_xlabel('Scale Factors')
ax.set_ylabel('Execution Time(Seconds)')
ax.set_title('Throughput 1 and 2')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# autolabel
autolabel(data1, ax,fontsize=5)
autolabel(data2, ax,fontsize=5)

plt.tight_layout()
plt.show()
```
