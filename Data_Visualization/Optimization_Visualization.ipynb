```python
import matplotlib.pyplot as plt
import numpy as np

# 生成数据
#x = np.arange(1, 11)  # x值：1到10
#y1 = np.random.randint(1, 10, size=10)  # 一个范围在1到10之间的数据集
#y2 = np.random.randint(1000, 10000, size=10)  # 一个范围在1000到10000之间的数据集

plt.figure(figsize=(10,6))

# 示例数据
x = ['1','3', '5', '7']
# 读取Excel文件
df = pd.read_excel("Optimization.xlsx", engine='openpyxl')
y1=df[df['Key'] == 'query_4.sql'].iloc[0].tolist()
first_element1 = y1.pop(0)  # pop()方法会返回被移除的元素，并且从原列表中移除它


df = pd.read_excel("origin.xlsx", engine='openpyxl')
y2=df[df['Key'] == 'query_4.sql'].iloc[0].tolist()


# 提取并打印列表的第一个元素
first_element2 = y2.pop(0)  # pop()方法会返回被移除的元素，并且从原列表中移除它
print("First Element:", first_element2)

# 使用对数刻度绘制Y轴
plt.yscale("log")

plt.plot(x, y1, 'o-', label="Optimization", color='blue')
plt.plot(x, y2, 's-', label="Origin", color='red')

plt.title("Comparison of Origin and Optimization of"+" "+ first_element2)
plt.xlabel("Scale Factor")
plt.ylabel("Exectution Time (Log Scale)")
plt.legend()
plt.grid(True, which="both", ls="--", c='0.65')

plt.show()
```
